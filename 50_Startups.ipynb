```python
df=df.loc[:,df.apply(pd.Series.nunique)!=1]
print(df.describe())
```

```python
df["State"].unique()
```

```python
df.sort_values('R&D Spend')
```

```python
ohe= OneHotEncoder()
ohe.fit_transform(df[["State"]]).toarray()

```

```python
s_array=ohe.fit_transform(df[["State"]]).toarray()
ohe.categories_[0]
```

```python
labels=ohe.categories_
pd.DataFrame(s_array, columns = labels)
```

```python
sns.countplot(data=df, x='State')
```

```python
# Melt the dataset for plotting boxplots and swarmplots
df_melted = pd.melt(df, value_vars=['R&D Spend', 'Administration', 'Marketing Spend', 'profit'])

# Create a boxplot and swarmplot to visualize the distribution of the numerical variables
sns.boxplot(x='variable', y='value', data=df_melted)
sns.swarmplot(x='variable', y='value', data=df_melted, size=2.8, color='black')
```

```python
features = pd.DataFrame(s_array, columns = labels)
#Merging both Dataframes
pd.concat([df, features], axis =1)
df1=pd.concat([df, features], axis =1)
#Removing 'State' column
df1.drop('State', axis=1)
```

```python
df1['profit']=df1.pop('profit')
df1
```

```python
df1.to_csv("Preprocessed_data_of_startups.csv", index=False)
```
